```python
# Parte 1: entrada de dados do usuario

nome = input("Qual o seu nome? \n")

#entrada das categorias de curso preferidas
Categorias = ["Wordpress",
              "Criação de Site",
              "E-commerce",
              "Marketing Digital",
              "Vendas",
              "Google Ads",
              "Design",
              "Negócios",
              "Youtube",
              "Criação de Conteúdo",
              "Programação"]
print('\033[1m' +"\n• Escolha o numero das categorias que mais lhe interessa\n" + '\033[0m')
for i in range(len(Categorias)): # mostrar tabela de categorias
   print(f"{i}. {Categorias[i]}")
  
categoria_preferida = (input("\nDigite o valor sem ponto e separado por vírgula:\n"))

#converter os dados obtidos como string em uma lista numerica
categoria_preferida = categoria_preferida.split(",")
pcategoria = []
for elemento in categoria_preferida:
  unid = int(elemento.strip())
  pcategoria.append(unid)


  
# entrada do tempo maximo disponivel para o curso
tempo_max = float(input("\nQuanto tempo voce tem pra se dedicar em um novo curso (em Min):\n"))


# Descobrir se o user quer ver apenas os gratuitos ou os pagos tambem
gratuidade = None # variavel boleana para fazer operações
# esse loop serve pra evitar que o user coloque um dado invalido como bool
while gratuidade != True and gratuidade != False:
  gratuidade = (input("\nMostrar apenas gratuitos? [S/N]\n"))
  if gratuidade == "N" or gratuidade == "n":
    gratuidade = False
  elif gratuidade == "S" or gratuidade == "s":
    gratuidade = True
  else:
    gratuidade = None
    print('opçao invalidada! Por favor, digite "S" ou "N"] ')

# Parte 2: Entrada dos dados dos cursos
nome_do_curso = ["Introdução ao Wordpress", 
                 "Criador de Sites", 
                 "Introdução ao WooCommerce", 
                 "Afiliado de Sucesso", 
                 "Venda mais com WhatsApp Business e Google Meu Negócio",
                 "Introdução ao Google Ads", 
                 "E-commerce e Vendas Online", 
                 "Canva: Design Fácil para Empreendedores", 
                 "Youtube para Iniciantes: como criar e crescer o seu canal",
                 "Rede de Display e Youtube", 
                 "Desenvolvimento de Plugins para WordPress"]
tempo_do_curso = [95, 25, 35, 85, 75, 150, 140, 120, 85, 335, 160 ]
categoria_do_curso = [
                      ["Wordpress", "Criação de Sites"], 
                      "Criação de Sites", 
                      ["E-commerce", "Criação de Sites", "Wordpress"], 
                      ["Marketing Digital", "Vendas"], 
                      [	"Marketing Digital", "Vendas", "Google Ads", "Negócios"], 
                      ["Marketing Digital", "Google Ads"], 
                      ["E-commerce", "Vendas", "Negócios"],
                      ["Design", "Negócios"], 
                      ["Youtube", "Criação de Conteúdo"], 
                      ["Marketing Digital", "Google Ads", "Youtube"],
                      ["Programação", "Wordpress"]]
curso_gratuiro = [True, True, True,	True,	True,	True,	True,	True,	True, False, True]
gratuito = [] # lista de strings para mostrar ao usuario se é pago ou free

for grat in range(len(curso_gratuiro)):
  if curso_gratuiro[grat] == True:
    gratuito.insert(grat, "Gratuito")
  elif curso_gratuiro[grat] == False:
    gratuito.insert(grat, "Pago")


# Parte 3: Relação entre dados do usuario e dados dos cursos
print(f'\n_________________________________________________________________________')
print(f'\nOlá {nome}, com base no seu perfil, te recomendamos os seguintes cursos: ')

cursos_recomendados = []
numero_de_recomendações = len(pcategoria) - 1
while numero_de_recomendações >= 0:
  print(f'\n{Categorias[pcategoria[numero_de_recomendações]].upper()}:\n')
  for i in range(len(categoria_do_curso)):
    if Categorias[pcategoria[numero_de_recomendações]] in categoria_do_curso[i] and tempo_do_curso[i] <= tempo_max and gratuidade == True and curso_gratuiro[i] == True:
      print(f'- {nome_do_curso[i]} | Duranção: {tempo_do_curso[i]} | {gratuito[i]} ')
      cursos_recomendados.append(nome_do_curso[i])
    elif Categorias[pcategoria[numero_de_recomendações]] in categoria_do_curso[i] and tempo_do_curso[i] <= tempo_max and gratuidade == False:
      print(f'- {nome_do_curso[i]} | Duranção: {tempo_do_curso[i]} | {gratuito[i]} ')
      cursos_recomendados.append(nome_do_curso[i])
  numero_de_recomendações -= 1


# Parte 4 Plus de analisar os cursos que tem mais de uma categoria selecionada para mostrar como prioridade
cursos_repetidos = []
cursos_recomendados.sort()  
print(f'\n\n!!!IMPORTANTE!!!\nSe você estiver em duvida de qual escolher, os curso abaixo tem o DOBRO de chance de lhe agradar:\n')
for cont in range(len(cursos_recomendados)):
  #print(cursos_recomendados.count(cursos_recomendados[cont]))
  if cursos_recomendados.count(cursos_recomendados[cont]) > 1:
    cursos_repetidos.append(cursos_recomendados[cont])
    cursos_repetidos = list(set(cursos_repetidos))
    print(f'- {nome_do_curso[cont]} | Duranção: {tempo_do_curso[cont]} | {gratuito[cont]} ')
```
